```python
import numpy as np
import pickle
```

```python
with open('train_qa.txt','rb') as f:
  train_data = pickle.load(f)
  
```

```python
with open('test_qa.txt','rb') as f:
  test_data = pickle.load(f)
```

```python
print(len(train_data),len(test_data))
```

```python
train_data[0][1]
```

```python
# Getting Vocabulary by combining train and test
train_test_data = train_data+test_data
```

```python
len(train_test_data)
```

```python
vocab = set()
#Building vocal
for story,ques,ans in train_test_data:
  vocab = vocab.union(set(story))
  vocab = vocab.union(set(ques))
vocab.add('no')
vocab.add('yes')
```

```python
len(vocab)
```

```python
# for keras pad sequences placeholder
vocab_len = len(vocab) + 1
```

```python
# Calculating longest story len
max_story_len = max([len(tuples[0]) for  tuples in train_test_data])
print(max_story_len)
```

```python
# Calculating longest question len
max_ques_len = max([len(tuples[1]) for  tuples in train_test_data])
print(max_ques_len)
```

```python
from keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(filters=[])
tokenizer.fit_on_texts(vocab)
```

```python
' '.join(train_data[0][0])
```

```python
def vectorizer(data_tuple,word_indexer = tokenizer.word_index, max_story_len = max_story_len,max_ques_len=max_ques_len):
  St = []
  Qu = []
  Ans = []
  for story,ques,answer in data_tuple:
    s = [word_indexer[word.lower()] for word in story]
    q = [word_indexer[word.lower()] for word in ques]
    a = np.zeros(len(word_indexer)+1)
    a[word_indexer[answer]] = 1
    St.append(s)
    Qu.append(q)
    Ans.append(a)
  return (pad_sequences(St,maxlen=max_story_len),pad_sequences(Qu,maxlen=max_ques_len),np.array(Ans))
      
```

```python
train_story_vec ,train_ques_vec ,train_ans_vec= vectorizer(train_data)
```

```python
test_story_vec ,test_ques_vec ,test_ans_vec= vectorizer(test_data)
```

```python
# Model Building
```

```python
from keras.models import Sequential,Model
from keras.layers.embeddings import Embedding
from keras.layers import LSTM,Input,Activation,Dense,Permute,Dropout,add,dot,concatenate
```

```python
# Input Placeholders
input_sequence = Input((max_story_len,))
question = Input((max_ques_len,))
```

```python
# Creating Encoders as given in the paper
```

```python
# Input Encoder M
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim=vocab_len,output_dim=64))
input_encoder_m.add(Dropout(0.3))
```

```python
# Input Encoder C
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim=vocab_len,output_dim=max_ques_len))
input_encoder_c.add(Dropout(0.3))
```

```python
# Input Encoder M
question_encoder = Sequential()
question_encoder.add(Embedding(input_dim=vocab_len,output_dim=64,input_length=max_ques_len))
question_encoder.add(Dropout(0.3))
```

```python
# Encoded Values
```

```python
encoded_m = input_encoder_m(input_sequence)
encoded_c = input_encoder_c(input_sequence)
encoded_question = question_encoder(question)
```

```python
dot_input_ques = dot([encoded_m,encoded_question],axes=(2,2))
dot_input_ques = Activation('softmax')(dot_input_ques)
```

```python
add_input_output = add([dot_input_ques,encoded_c])
add_input_output = Permute((2,1))(add_input_output)
```

```python
answer = concatenate([add_input_output,encoded_question])
```

```python
answer
```

```python
answer = LSTM(32)(answer)
answer = Dropout(0.5)(answer)
answer = Dense(vocab_len)(answer)
```

```python
answer = Activation('softmax')(answer)
```

```python
model = Model([input_sequence,question],answer)
```

```python
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit([train_story_vec,train_ques_vec],train_ans_vec,batch_size=32,epochs=100,validation_data=([test_story_vec,test_ques_vec],test_ans_vec))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(model.history.history['acc'])
plt.plot(model.history.history['val_acc'])
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.legend(['train','test'])
```

```python
def test_vectorizer(sample_story,query,word_indexer = tokenizer.word_index, max_story_len = max_story_len,max_ques_len=max_ques_len):
  data_tuple = [(sample_story.split(),query.split())]
  St = []
  Qu = []
  for story,ques in data_tuple:
    s = [word_indexer[word.lower()] for word in story]
    q = [word_indexer[word.lower()] for word in ques]
    St.append(s)
    Qu.append(q)
  return (pad_sequences(St,maxlen=max_story_len),pad_sequences(Qu,maxlen=max_ques_len))
```

```python
sample_story = "John left the kitchen . Mary dropped the football in the garden . John went to the hallway ."
```

```python
query = "Is John in the Hallway ?"
```

```python
sample_story,query = test_vectorizer(sample_story,query)
```

```python
pred = model.predict([sample_story,query])
```

```python
for k,v in tokenizer.word_index.items():
  if v == np.argmax(pred):
    print(k)
```
